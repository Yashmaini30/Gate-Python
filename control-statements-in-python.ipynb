```python
# match-case
# match expression :
#     case object :
#         statement(s)
#     case object :
#         statement(s).......

# 1. break is not req
# 2. default case is optional
# 3. to write empty-cases , use "pass" statement
# 4. Objects can be of any fundamental datatype
# 5. object can be in any random order
# 6. Duplicates cases are allowed
# 7. cases can be mixed type
```

```python
i = int(input())
match i:
    case -2:
        print("yellow")
    case -1:
        print("red")
    case 'd':
        print("green")
    case "hello":
        print("blue")
    case _:
        print("No match")
```
