```python
# Dictionaries
# ordered collection
# only duplicated values, not keys
# only values are mutable
# coll of key-value pair

d={'a':1,'b':2,'c':3}
print(d.keys())
print(d.values())
print(d.items())

for (i,j) in d.items():
    print(i,j)
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b  
    return a

print(gcd(8, 100))

def rGcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)

print(rGcd(8, 100))
```
