```python
s = "Gate"
print(s.upper())         # GATE
print(s.lower())         # gate
print(s.isnumeric())     # False
print(s.isalpha())       # True
print(s.islower())       # False
print(s.isupper())       # False
print(s.center)          # <built-in method center of str object at 0x...>
print(s.center(10))      # "   Gate   "
print(s.center(10, '#')) # "###Gate###"
print(s.isascii())       
```

```python
# TUPLE
# ordered collection 
# supp both pos and neg
# supp duplicates
# supp diff type of data
# immutable
# tuples can be modified as below:
    # 1. covert to list
    # 2. modify
    # 3. convert back to tuple
# but support, clear() and del keyword
# nested tuples 
```

```python
# unpacking tuple
t=(10,20,30)
(a,b,c)=t
print(a,b,c)

t1=(10,20,30,40,60)
a,b,*c=t
print(a,b,c)

t2 = (1, 2, 3, 4, 5)
x, *y, z = t2
print(x) 
print(y)  
print(z)  
```

```python
# # mthod of tuples
# index() -> return first occ of given ele
# count() -> fre of ele

t=(10,20,30,40,10)
i=t.index(30)
j=t.count(10)
print(i+j)
```

```python
# sets
# undordered(no index)
# no duplicates
# Immutable (add and remove items)
t=tuple([3,4,5])
print(type(t))

s={10,20,30,10,50}
print(s)

s={}
print(type(s))

s=set()
print(type(s))
```
