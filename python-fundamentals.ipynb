```python
# Integers: A number without a fractional part
# 1. Decimal integer
# 2. Octal integer
# 3. Binary Integer
# 4. HexaDecimal integer

a = 43
b = hex(a)
c = bin(a)
d = oct(a)

print(a,b,c,d,sep="\n")

i= 43 + 0x2b + 0b101011 + 0o53 #all numbers need to be converted to decimal  
print(i)
```

```python
# Python supports Unicode Format
# [A:65,a:97,space:32,0:48]

# ord() function
# It returns Unicode value of the respective char

print(ord('A'))
print(ord('3'))
print(ord('d'))

#chr function
# it returns Unicode character of given value

print(chr(65))
print(chr(100))
print(chr(51))

print(ord('t')+ord('B')-ord('8'))
```
