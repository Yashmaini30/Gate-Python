```python
count = 1
i = 0

while i <= 10:
    match i:
        case 3:
            count += 2
        case 1:
            count += 4
        case 7:
            count *= 2
        case 0:
            count -= 2
        case 6:
            if count != 0:
                count //= 2
        case 9:
            count += 3
        case _:
            count -= 1 
    i=i+2

print(count)
```

```python
count=0
i=1
while i<=5:
    count = count -2
    i=i+2
print(count)
```

```python
res=1
i=1
j=5
while i<5:
    res=res+i
    while j>i:
        res=res+j
        j=j-2
    i=i+2
print(res)
```

```python
ans=1
i=2
for j in range(3,7):
    ans += i+j
    while i<ans:
        j+=1
        i+=j
print(ans)
```

```python
res=1
for i in range(2,7,3):
    for j in range(i):
        res += i+j
print(res)
```

```python
# # jumping control statement 
# 1. break - move out of the current loop
# 2. continue - move to the beginning of the next iteration of respective loop
# 3. pass - a placeholder statement

for i in range(5):
    if i==3:
        break 
    print(i)
```

```python
for i in range(7):
    if i==3:
        continue
    print(i)
```

```python
for i in range(1,7,2):
    for j in range (i):
        print(i+j)
        if j%2==1:
            break
```

```python
for i in range(1,7,2):
    for j in range(i):
        print(i+j)
        if j%2==1:
            continue
```
